```python
df = events_df
df['Start Date/Time'] = pd.to_datetime(df['Start Date/Time'])
df['End Date/Time'] = pd.to_datetime(df['End Date/Time'])
```

```python
def early_late(df):

    # Earliest and latest dates in df
    event_start_earliest = df['Start Date/Time'].min()
    event_start_latest = df['Start Date/Time'].max()
    event_end_earliest = df['End Date/Time'].min()
    event_end_latest = df['End Date/Time'].max()

    print(f"Earliest Start Date/Time: {event_start_earliest}")
    print(f"Earliest End Date/Time: {event_end_earliest}")
    print(f"Latest Start Date/Time: {event_start_latest}")
    print(f"Latest End Date/Time: {event_end_latest}")

early_late(events_df)
```

```python
def reformat(df):
    # Round datetime to the nearest hour
    df['Start Date/Time'] = df['Start Date/Time'].dt.round('h')
    df['End Date/Time'] = df['End Date/Time'].dt.round('h')


    # Extract the required time components
    df['datetime_formatted'] = df['datetime'].dt.strftime('%Y-%m-%d-%H')
    df['hour'] = df['datetime'].dt.hour
    df['day_of_week'] = df['datetime'].dt.dayofweek
    df['week'] = df['datetime'].dt.isocalendar().week
    df['month'] = df['datetime'].dt.month - 1  # Convert to 0-11 for Jan-Dec
    df['day_of_month'] = df['datetime'].dt.day
    df['year_month'] = df['datetime'].dt.to_period('M').astype(str)
    
    return df
```
