```python
import requests
import pandas as pd
import openai
import os
```

```python
base_url = "https://data.cityofnewyork.us/resource/nyis-y4yr.json"

# Parameters for the API request
params = {
    "$limit": 1000,  # Number of rows per request
    "$offset": 0     # Starting point for each request
}

# List to store the data
data = []
```

```python
# Loop to fetch the data in chunks of 1000
while True:
    # Make the API request
    response = requests.get(base_url, params=params)
    
    # Check if the request was successful
    if response.status_code != 200:
        print("Error fetching data: ", response.status_code)
        break

    # Parse the response JSON
    chunk = response.json()
    
    # If the chunk is empty, we've fetched all the data
    if not chunk:
        break
    
    # Add the chunk to our data list
    data.extend(chunk)
    
    # Update the offset for the next request
    params["$offset"] += 1000

    # Optional: Print the progress
    print(f"Fetched {params['$offset']} rows")
```
